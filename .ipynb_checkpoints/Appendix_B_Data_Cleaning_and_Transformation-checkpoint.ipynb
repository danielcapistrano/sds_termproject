```python
import pandas as pd
df = pd.read_csv("./data/raw/Base.csv")
```

```python
columns_to_drop = ['payment_type', 'employment_status', 'housing_status']
df = df.drop(columns=columns_to_drop)
```

```python
df
```

```python
unique_sources = df['source'].unique()
for category in unique_sources:
    df[category] = df['source'] == category
df = df.drop(columns = 'source')
```

```python
unique_device_os = df['device_os'].unique()
for category in unique_device_os:
    df[category] = df['device_os'] == category
df = df.drop(columns = 'device_os')
```

```python
df.dtypes
```

```python
df.to_csv('./data/raw/Base_cleaned.csv')
```
