```python
import pandas as pd
df = pd.read_csv("./data/raw/Base.csv")
```

```python
# Dropping anonymized columns
columns_to_drop = ['payment_type', 'employment_status', 'housing_status']
df = df.drop(columns=columns_to_drop)
```

```python
df.describe()
```

```python
# The column 'device_fraud_count' contains only the value 0 for all cases (constant)
df = df.drop(columns='device_fraud_count')
```

```python
df['is_internet'] = df['source'] == 'INTERNET'
df = df.drop(columns = 'source')
```

```python
df['is_windows_or_mac'] = df['device_os'].isin(['windows', 'macintosh'])
df = df.drop(columns = 'device_os')
```

```python
df.dtypes
```

```python
df.to_csv('./data/raw/Base_cleaned.csv', index=False)
```
