```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import random
p = 0.1  # 10% of the lines
# keep the header, then take only 1% of lines
# if random from [0,1] interval is greater than 0.01 the row will be skipped
df = pd.read_csv(
         "CreditCard.csv",
         header=0,
         skiprows=lambda i: i>0 and random.random() > p
)
df.describe()
```

```python
import pandas as pd

# Separate majority and minority classes
majority_class = df[df['fraud_bool'] == 0]
minority_class = df[df['fraud_bool'] == 1]

# Determine the number of instances in the classes
majority_count = len(majority_class)
minority_count = len(minority_class)

print(f'majority count: {majority_count}')
print(f'minority count: {minority_count}')

print(f'minor:major ration: {minority_count/majority_count}')
```

```python
# Undersample majority class to match minority class
undersampled_majority = majority_class.sample(n=minority_count)

# Combine minority and undersampled majority classes
balanced_df = pd.concat([undersampled_majority, minority_class])

# Shuffle the balanced dataset to ensure randomness
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Optional: Check the class distribution after undersampling
print(balanced_df['fraud_bool'].value_counts())
```
